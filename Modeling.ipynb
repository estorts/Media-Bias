```python
def convert_to_num(target):
    if target == 'left':
        return 0
    elif target == 'left-center':
        return 1
    elif target == 'center':
        return 2
    elif target == 'right':
        return 3
    elif target == 'right-center':
        return 4
    else:
        return np.nan
```

```python
train_data, test_data = train_test_split(df, test_size=0.25, random_state=42)
```
